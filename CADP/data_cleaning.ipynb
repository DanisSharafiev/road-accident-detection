```python
import re
import os
```

```python
mask = r'.*_\d+s\.jpg'
pattern = re.compile(mask)
```

```python
filenames = os.listdir('Accident/')
filenames[:5]
```

```python
filegroup = {} # filename : indexes
for name in filenames:
    if pattern.match(name):
        filename_without_index = re.sub(r'_(\d+)s\.jpg$', '', name)
        index = re.search(r'_(\d+)s\.jpg$', name).group(1)
        if filename_without_index in filegroup:
            filegroup[filename_without_index].append(index)
        else:
            filegroup[filename_without_index] = [index]

for key in filegroup:
    filegroup[key].sort()

print(f'Total unique file groups: {len(filegroup)}')
```

```python
print(filegroup)
```

```python
# Median coef shows how many images to keep from the center of each group
MEDIAN_COEF = 0.1
# Shift coef shows how much shift from the center to the right
SHIFT_COEF = 0.8

os.makedirs('Cleaned_Accident', exist_ok=True)
```

```python
for key, value in filegroup.items():
    total_images = len(value)
    images_to_keep = max(1, int(total_images * MEDIAN_COEF))
    center_index = int(total_images * SHIFT_COEF)
    start_index = max(0, center_index - images_to_keep // 2)
    end_index = min(total_images, start_index + images_to_keep)
    indexes_to_keep = set(value[start_index:end_index])
    
    # copy into Cleaned_Accident only the selected images
    for index in value:
        if index in indexes_to_keep:
            src_path = os.path.join('Accident', f'{key}_{index}s.jpg')
            dst_path = os.path.join('Cleaned_Accident', f'{key}_{index}s.jpg')
            os.system(f'cp "{src_path}" "{dst_path}"')
```
